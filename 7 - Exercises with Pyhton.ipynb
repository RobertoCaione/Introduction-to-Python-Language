#### FFT Fast Fourier Transform

````python
# Compute the one-dimensional discrete Fourier Transform

import numpy as np
import matplotlib.pyplot as plt

t = np.arange(100)
sp = np.fft.fft(np.sin(t))
freq = np.fft.fftfreq(t.shape[-1])

# plot the results 
fig,ax=plt.subplots(1,3,figsize=(15,5))
ax[0].plot(np.sin(t), color='yellow')
ax[0].plot(freq, color='red')
ax[0].set_title('sin function and freq')
ax[1].plot(sp.real, color ='blue')
ax[1].set_title('real part')
ax[2].plot(sp.imag, color='green')
ax[2].set_title('imaginary part')
plt.show()
````

### $\chi^2$ TEST

The chi-square distribution is used in the common chi-square tests for goodness of fit of an observed. From the theory:

$\chi^2=\sum_i \frac{(f_{observed}-f_{expected})^2}{f_{expected}}$ 

````python
from scipy.stats import chisquare  # import useful package
````

When just $f_{observed}$ is given, it is assumed that $f_{expected}$ are uniform and given by the mean of the observed frequencies.

````python
chisquare([16, 18, 16, 14, 12, 12])
````

In addition to $f_{observed}$ also $f_{expected}$ can be given.

````python
chisquare([16, 18, 16, 14, 12, 12], f_exp = [16, 16, 16, 16, 16, 16])
````

If $f_{observed}$ is 2-D, by default the test is applied to each column.

````python
import numpy as np
obs = np.array([[16, 18, 16, 14, 12, 12], [16, 18, 16, 14, 12, 12]]).T
print(obs.shape)
chisquare(obs)
````

````python
obs = np.array([[16, 18, 16, 14, 12, 12], [16, 18, 16, 14, 12, 12], [16, 18, 16, 14, 12, 12]]).T
print(obs.shape)
chisquare(obs)
````

By setting axis=None, the test is applied to all data in the array, which is equivalent to apply the test to the flattened array.

````python
chisquare(obs, axis=None)
````

ddof is the change to make to the default degrees of freedom.

````python
chisquare([16, 18, 16, 14, 12, 12], ddof=1)
````

The calculation of the p-values is done by broadcasting the chi-squared statistic with ddof.

````python
chisquare([16, 18, 16, 14, 12, 12], ddof=[0,1,2])
````

$f_{observed}$ and $f_{expected}$ are also broadcast. In the following, $f_{obs}$ has shape (6) and $f_{exp}$ has shape (2, 6), so the result of broadcasting $f_{observed}$ and $f_{expected}$ has shape (2, 6). To compute the desired chi-squared statistics, we use axis=1:

````python
chisquare([16, 18, 16, 14, 12, 12], f_exp=[[16, 16, 16, 16, 16, 8], [8, 20, 20, 16, 12, 12]], axis=1)
````

#### From Mpc to arcsec

````python
from astropy import cosmology               # import useful package
cosmo = cosmology.FlatLambdaCDM(70, 0.26)   # setting cosmology
zl = 1.0                                    # setting redshift
rs_mpc = 2.0                                # setting scale

# define a function able to convert Mpc in arcsec given the redshift 
# ----> from observed distance to real distance
def mpc_arcsec(mpc,zl):                             
    dz = cosmo.angular_diameter_distance(zl)        # angular diameter distance
    rs_mpc = mpc
    rs_rad = rs_mpc / dz                            # scale radius (rad)
    rs_deg = rs_rad*(180/np.pi)                     # scale radius (deg)
    rs_arcsec = rs_deg*3600                         # scale radius (arcsec)
    return rs_arcsec

rs_arcsec = mpc_arcsec(rs_mpc,zl)
print rs_arcsec.value
````

#### RA DEC in sexa to RA DEC in deg

#### Distance modulus 
The distance modulus is a way of expressing distances that are often used in astronomy. It describes distances on a logarithmic scale based on the astronomical magnitude system. The distance modulus $\mu = m  M$ is the difference between the apparent magnitude $m$ (ideally, corrected from the effects of interstellar absorption) and the absolute magnitude $M$ of an astronomical object.

````python
from astropy import cosmology               # import useful package
cosmo = cosmology.FlatLambdaCDM(70, 0.26)   # setting cosmology
z = 0.5
print(cosmo.distmod(z))
````

````python
from astropy import cosmology               # import useful package
import numpy as np

z = 0.5
dl = cosmo.luminosity_distance(z)
print(dl)
dl = dl.to('parsec').value
logd = np.log10(dl)
mu = 5*logd-5
print(mu)
````

#### SHAPELY to estimate area of figure and intersection region

````python
import shapely 
from shapely.ops import unary_union, polygonize
from shapely.geometry import Point, LineString, Polygon
import numpy as np
import random
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt

circles = [Point(random.random(), random.random()).buffer(random.random() * 0.1) for i in range(15)]

rings = [LineString(list(pol.exterior.coords)) for pol in circles]
union = unary_union(rings)
result = [geom for geom in polygonize(union)]

print(len(result))
fig, ax = plt.subplots(figsize=(8, 8))
for polygon in result:
    poly = Polygon(np.array(polygon.exterior), facecolor='g', edgecolor='k', lw=1, alpha=0.4)
    ax.add_patch(poly)
ax.relim()
ax.autoscale()

aree = np.array([])
for i in range(len(result)):
    aree = np.append(aree,result[i].area)
print(aree)
````
